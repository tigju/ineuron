Box tuple is a tuple with four values, representing image coordinates. First two integers is x and y of top left conner, second two integers is x and y of bottom right conner. Also box tuple represent any rectangular region in an image.

##### 8. Image objects do not have drawing methods. What kind of object does? How do you get this kind of object?
ImageDraw object has drawing methods like line(), rectangle(), ellipse(), polygon(). To get this object you have to pass image object to ImageDraw.Draw(), and then apply drawing methods to it.


```python
from PIL import ImageDraw
im = Image.open("modified.jpg")
draw = ImageDraw.Draw(im)
```

```python
draw.line([(0, 0), (99, 0), (99, 99), (0, 99), (0, 0)], fill='red') 
draw.rectangle((5, 10, 25, 25), fill='blue')
draw.ellipse((30, 30, 50, 60), fill='red')
draw.polygon(((50, 57), (69, 62), (70, 75), (80, 60), (80, 99)), fill='pink')
```
